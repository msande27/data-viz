```python
%matplotlib inline

import pandas as pd
import matplotlib

matplotlib.style.use('ggplot')

hotdog_dir = r"/Users/maxmealy/Documents/JHU AAP/Data Visualization/Week 3/hotdogs.csv"
```

```python
# hotdogs <- read.csv("hotdogs.csv")
hotdogs = pd.read_csv(hotdog_dir, index_col='Year')

#head(hotdogs)
hotdogs.head(n=5)
```

```python
#dim(hotdogs)
hotdogs.shape[1]
```

```python
#barplot(hotdogs$Dogs.eaten)
hotdogs['Dogs eaten'].plot(kind='bar')
```

```python
#barplot(hotdogs$Dogs.eaten, names.arg=hotdogs$Year, xlab="Year", ylab="Hot dogs and buns eaten")
hotdogbar = hotdogs['Dogs eaten'].plot(kind='bar')
hotdogbar.set(xlabel = "Year", ylabel = "Hot dogs and buns eaten")
```

```python
#barplot(hotdogs$Dogs.eaten, names.arg=hotdogs$Year, col="#475BBF", border=NA, xlab="Year", ylab="Hot dogs and buns eaten")
hotdogbar = hotdogs['Dogs eaten'].plot(kind='bar', color ="#475BBF")
hotdogbar.set(xlabel = "Year", ylabel = "Hot dogs and buns eaten")
hotdogbar.patch.set_visible(False)
```

```python
#hotdogs$fill_colors <- ""
#hotdogs$fill_colors[which(hotdogs$Country == "United States")] <- "#821122"
#hotdogs$fill_colors[which(hotdogs$Country != "United States")] <- "#cccccc"
#barplot(hotdogs$Dogs.eaten, names.arg=hotdogs$Year, col=hotdogs$fill_colors, border=NA, xlab="Year", ylab="Hot dogs and buns (HDB) eaten")
hotdogs['fill_colors'] = ""
hotdogs.loc[hotdogs['Country'] == "United States", 'fill_colors'] = "#821122"
hotdogs.loc[hotdogs['Country'] != "United States", 'fill_colors'] = "#cccccc"
hotdogbar = hotdogs['Dogs eaten'].plot(kind='bar', color =hotdogs['fill_colors'])
hotdogbar.set(xlabel = "Year", ylabel = "Hot dogs and buns eaten")
hotdogbar.patch.set_visible(False)
```

```python
#hotdogs$fill_colors[which(hotdogs$Winner == "Joey Chestnut ")] <- "#54278f"
#hotdogs$fill_colors[which(hotdogs$Winner != "Joey Chestnut ")] <- "#cccccc"
#barplot(hotdogs$Dogs.eaten, names.arg=hotdogs$Year, col=hotdogs$fill_colors, border=NA, xlab="Year", ylab="Hot dogs and buns (HDB) eaten")
hotdogs.loc[hotdogs['Winner']=="Joey Chestnut ",'fill_colors'] = "#54278f"
hotdogs.loc[hotdogs['Winner']!="Joey Chestnut ",'fill_colors'] = "#cccccc"
hotdogbar = hotdogs['Dogs eaten'].plot(kind='bar', color =hotdogs['fill_colors'])
hotdogbar.set(xlabel = "Year", ylabel = "Hot dogs and buns (HDB) eaten")
hotdogbar.patch.set_visible(False)
```

```python
#hotdogs$fill_colors <- "#cccccc"
#hotdogs$fill_colors[which(hotdogs$New.record == 1)] <- "#821122"
#barplot(hotdogs$Dogs.eaten, names.arg=hotdogs$Year, col=hotdogs$fill_colors, border=NA, xlab="Year", ylab="Hot dogs and buns (HDB) eaten")
hotdogs['fill_colors'] = "#cccccc"
hotdogs.loc[hotdogs['New record']==1,'fill_colors'] = "#821122"
hotdogbar = hotdogs['Dogs eaten'].plot(kind='bar', color =hotdogs['fill_colors'])
hotdogbar.set(xlabel = "Year", ylabel = "Hot dogs and buns (HDB) eaten")
hotdogbar.patch.set_visible(False)
```

```python
#barplot(hotdogs$Dogs.eaten, names.arg=hotdogs$Year, col=hotdogs$fill_colors, border=NA, space=0.5, xlab="Year", ylab="Hot dogs and buns (HDB) eaten")
hotdogbar = hotdogs['Dogs eaten'].plot(kind='bar', color =hotdogs['fill_colors'], width=.5)
hotdogbar.set(xlabel = "Year", ylabel = "Hot dogs and buns (HDB) eaten")
hotdogbar.patch.set_visible(False)
```

```python
#barplot(hotdogs$Dogs.eaten, names.arg=hotdogs$Year, col=hotdogs$fill_colors, border=NA, space=0.25, xlab="Year", ylab="Hot dogs and buns (HDB) eaten", main="Nathan's Hot Dog Eating Contest Results, 1980-2010")
hotdogbar = hotdogs['Dogs eaten'].plot(kind='bar', color =hotdogs['fill_colors'], width=.75, title="Nathan's Hot Dog Eating Contest Results, 1980-2010\n")
hotdogbar.set(xlabel = "Year", ylabel = "Hot dogs and buns (HDB) eaten")
hotdogbar.patch.set_visible(False)
```
